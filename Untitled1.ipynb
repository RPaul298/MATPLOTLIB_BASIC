```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from datetime import datetime
```

```python
data=pd.read_csv("householdtask3.csv") 
print(data.head(10)) 
```

```python
#Visualizing using line chart
plt.plot(data['year'])
plt.plot(data['own'])

plt.title('Line Chart')

plt.xlabel('year')
plt.ylabel('own')

plt.show()
```

```python
#Visualizing using Bar chart
plt.bar(data['year'], data['own'])

plt.title('Bar Chart')

plt.xlabel('year')
plt.ylabel('own')

plt.show()
```
